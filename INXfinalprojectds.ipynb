```python
import pandas as pd 
from pandas import Series,DataFrame
```

```python
data= pd.read_csv('INX_Future_Inc_Employee_Performance_CDS_Project2_Data_V1.8.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.corr().loc[:,"PerformanceRating"]     # pick the values having highest corr with rating
```

```python
from sklearn.preprocessing import LabelEncoder # labelencoder is used for converting text into digits
```

```python
enc=LabelEncoder()
```

```python
data.Gender              = enc.fit_transform(data.Gender)
data.EducationBackground = enc.fit_transform(data.EducationBackground)
data.MaritalStatus = enc.fit_transform(data.MaritalStatus)
data.EmpDepartment = enc.fit_transform(data.EmpDepartment)
data.EmpJobRole = enc.fit_transform(data.EmpJobRole)
data.BusinessTravelFrequency = enc.fit_transform(data.BusinessTravelFrequency)
data.EmpLastSalaryHikePercent = enc.fit_transform(data.EmpLastSalaryHikePercent)
data.EmpEnvironmentSatisfaction       = enc.fit_transform(data.EmpEnvironmentSatisfaction      )
```

```python
data.head(1)
```

```python
x= data.loc[:,['EmpEnvironmentSatisfaction','EmpLastSalaryHikePercent','EmpWorkLifeBalance']] # top3 factors affect performance
y= data.PerformanceRating
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.metrics import accuracy_score,classification_report
from sklearn.svm import SVC
model=SVC(kernel='rbf',C=6,gamma=0.2)
model.fit(x_train,y_train)
y_predict=model.predict(x_test)
print(accuracy_score(y_test,y_predict))
pd.crosstab(y_test,y_predict)
```
